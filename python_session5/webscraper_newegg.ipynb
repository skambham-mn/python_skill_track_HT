```python
from bs4 import BeautifulSoup as Soup
from urllib.request import urlopen as uReq
```

```python
page_url ="https://www.newegg.com/p/pl?N=100007671%20601306860&cm_sp=Cat_CPU-Processors_1-_-Visnav-_-Intel-CPU_2"
```

```python
uClient = uReq(page_url)
```

```python
page_soup = Soup(uClient.read(), "html.parser")
uClient.close()
```

```python
containers = page_soup.findAll("div", {"class":"item-container"})
```

```python
containers[0].img['title']
```

```python
containers[0].findAll("li", {"class":"price-current"})[0].text.strip()
```

```python
out_filename ="intel_processors.csv"
headers ="item_name, price\n"
```

```python
f =open(out_filename, "w")
f.write(headers)

for container in containers:
    item_name = container.img['title'].title()
    price = container.findAll("li", {"class":"price-current"})[0].text.strip()
    print("item_name:" + item_name + "\n")
    print("price:" + price + "\n")
    f.write(item_name + ", " + price + "\n")

f.close()
    
    
```
